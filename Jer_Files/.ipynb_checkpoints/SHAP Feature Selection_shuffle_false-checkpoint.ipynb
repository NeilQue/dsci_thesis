```python
class WindowGenerator():
    def __init__(self, input_width, label_width, shift, train_df, val_df, test_df, label_columns=None):
        # Store the raw data.
        self.train_df = train_df
        self.val_df = val_df
        self.test_df = test_df

        # Work out the label column indices.
        self.label_columns = label_columns
        if label_columns is not None:
            self.label_columns_indices = {name: i for i, name in enumerate(label_columns)}
        self.column_indices = {name: i for i, name in enumerate(train_df.columns)}

        # Work out the window parameters.
        self.input_width = input_width
        self.label_width = label_width
        self.shift = shift

        self.total_window_size = input_width + shift

        self.input_slice = slice(0, input_width)
        self.input_indices = np.arange(self.total_window_size)[self.input_slice]

        self.label_start = self.total_window_size - self.label_width
        self.labels_slice = slice(self.label_start, None)
        self.label_indices = np.arange(self.total_window_size)[self.labels_slice]

    def __repr__(self):
        return '\n'.join([
            f'Total window size: {self.total_window_size}',
            f'Input indices: {self.input_indices}',
            f'Label indices: {self.label_indices}',
            f'Label column name(s): {self.label_columns}'])

    def split_window(self, features):
        inputs = features[:, self.input_slice, :]
        labels = features[:, self.labels_slice, :]
        if self.label_columns is not None:
            labels = tf.stack([labels[:, :, self.column_indices[name]] for name in self.label_columns], axis=-1)

        # Slicing doesn't preserve static shape information, so set the shapes
        # manually. This way the `tf.data.Datasets` are easier to inspect.
        inputs.set_shape([None, self.input_width, None])
        labels.set_shape([None, self.label_width, None])

        return inputs, labels
    
    # Creating tf datasets for more convenient use and integration into model in the future
    def make_dataset(self, data, shuffle = True):
        data = np.array(data, dtype=np.float32)
        ds = tf.keras.utils.timeseries_dataset_from_array(
            data=data,
            targets=None,
            sequence_length=self.total_window_size,
            sequence_stride=1,
            shuffle=shuffle,
            batch_size=32,)

        ds = ds.map(self.split_window)

        return ds
    
    # properties to access them as tf datasets
    @property
    def train(self):
        return self.make_dataset(self.train_df)

    @property
    def val(self):
        return self.make_dataset(self.val_df)

    @property
    def test(self):
        return self.make_dataset(self.test_df, False)

    @property
    def example(self):
        """Get and cache an example batch of `inputs, labels` for plotting."""
        result = getattr(self, '_example', None)
        if result is None:
            # No example batch was found, so get one from the `.train` dataset
            result = next(iter(self.train))
            # And cache it for next time
            self._example = result
        return result
```

```python
def r_square(y_true, y_pred):
    x = y_true
    y = y_pred
    mx = K.mean(x, axis=0)
    my = K.mean(y, axis=0)
    xm, ym = x - mx, y - my
    r_num = K.square(K.sum(xm * ym))
    x_square_sum = K.sum(xm * xm)
    y_square_sum = K.sum(ym * ym)
    r_den = (x_square_sum * y_square_sum) + K.epsilon()
    
    r = r_num / r_den
    return r
```

```python
def NSE(y_true, y_pred):
    '''
    This is the Nash-Sutcliffe Efficiency Coefficient
    '''
    y_pred = K.flatten(y_pred)
    y_true = K.flatten(y_true)

    
    SS_res =  K.sum(K.square(y_true - y_pred)) 
    SS_tot = K.sum(K.square(y_true - K.mean(y_true))) 
    
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )
```

```python
# For easy compiling and fitting of different models
MAX_EPOCHS = 20

def compile_and_fit(model, window, patience=2):
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=patience, mode='min')

    model.compile(
        loss=tf.keras.losses.MeanSquaredError(), 
        optimizer='adam', 
        metrics=[tf.keras.metrics.MeanSquaredError(), NSE, r_square]
    )

    history = model.fit(
        window.train, 
        epochs=MAX_EPOCHS,
        validation_data=window.val,
        callbacks=[early_stopping]
    )

    return history
```

```python
# LSTM
lstm_model_sto_nino = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64, return_sequences=True),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=1)
])

lstm_model_montalban = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64, return_sequences=True),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=1)
])
```

```python
lstm_history_sto_nino = compile_and_fit(lstm_model_sto_nino, window_sto_nino)
```

```python
def rolling_forecast(model, dataset, steps=10):
    """Generates rolling forecasts step by step"""
    predictions = []
    for inputs, labels in dataset:
        pred = model.predict(inputs)  # Predict current timestep
        predictions.append(pred)
    return np.concatenate(predictions, axis=0)

test_preds = rolling_forecast(lstm_model_sto_nino, window_sto_nino.test)
```

```python
def get_predictions(model, test_set):
    y_pred = []
    
    for x, _ in test_set:
        predictions = model(x)
        for y in predictions:
            y_pred.append(y)
    
    y_pred = np.array(y_pred)
    y_pred = y_pred.reshape(-1)
    
    return y_pred
```

## SHAP Feature Selection (Sto. Nino)

```python
import random
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
import shap
import numpy as np
import matplotlib.pyplot as plt

# Get training data for the background distribution
train_data = []
for inputs, _ in window_sto_nino.train.take(5): 
    batch_data = inputs.numpy()
    train_data.append(batch_data)
train_data = np.concatenate(train_data, axis=0)  # Combine batches

# Flatten the training data for the explainer
train_data_flat = train_data.reshape(train_data.shape[0], -1)

# Get all test data
test_data = []
for inputs, _ in window_sto_nino.test.take(5):  # Take all test data
    batch_data = inputs.numpy()
    test_data.append(batch_data)
test_data = np.concatenate(test_data, axis=0)  # Combine batches

# Flatten the test data
test_data_flat = test_data.reshape(test_data.shape[0], -1)

# Prediction function that returns a single value (last time step of water level)
def predict_last_step(X):
    # Reshape back to 3D
    batch_size = X.shape[0]
    time_steps = window_sto_nino.input_width
    features = len(window_sto_nino.train_df.columns)
    
    X_reshaped = X.reshape(batch_size, time_steps, features)
    
    # Get model predictions
    preds = lstm_model_sto_nino.predict(X_reshaped, verbose=0)
    
    # Return the last time step prediction for water level
    return preds[:, -1, 0]  # Last time step, first feature (water level)

# Create the explainer using training data as background
explainer = shap.Explainer(predict_last_step, train_data_flat)  

# Calculate SHAP values for all test samples
shap_values = explainer(test_data_flat)  # Directly call explainer on test data

# Extract the SHAP values
shap_values_array = shap_values.values  # Get the actual SHAP values

```

```python
# Get feature names
feature_names = list(window_sto_nino.train_df.columns)

# Create feature name labels for each time step
time_step_feature_names = []
for t in range(window_sto_nino.input_width):
    for feature in feature_names:
        time_step_feature_names.append(f"{feature} (t-{window_sto_nino.input_width - t})")

# Create the summary plot with proper feature names
plt.figure(figsize=(12, 8))
shap.summary_plot(shap_values, test_data_flat, feature_names=time_step_feature_names)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get the number of features and time steps
num_features = len(feature_names)
num_time_steps = window_sto_nino.input_width

# Reshape SHAP values: (num_samples, total_features) → (num_samples, time_steps, num_features)
shap_values_reshaped = shap_values_array.reshape(test_data_flat.shape[0], num_time_steps, num_features)

# Aggregate SHAP values across time steps (mean absolute impact per feature)
shap_feature_importance = np.mean(np.abs(shap_values_reshaped), axis=(0, 1))  # Mean across samples & time

plt.figure(figsize=(12, 8))
bars = plt.bar(feature_names, shap_feature_importance, color='skyblue')
plt.title('Sto. Nino Water Level Feature Importance', fontsize=16)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Average |SHAP Value|', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.tight_layout()


# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.models import Model, clone_model

def create_lstm_ensemble_with_averaged_weights(model1, model2):
    """
    Create an ensemble LSTM model by averaging the weights of two pre-trained LSTM models.
    
    Args:
        model1: First pre-trained LSTM model
        model2: Second pre-trained LSTM model
        
    Returns:
        A new LSTM model with averaged weights
    """
    # Verify that the models have the same architecture
    if len(model1.layers) != len(model2.layers):
        raise ValueError("Models have different architectures (different number of layers)")
    
    # Clone one of the models to create the ensemble model structure
    ensemble_model = clone_model(model1)
    
    # Set the weights of the ensemble model as the average of the two models
    for i, (layer1, layer2) in enumerate(zip(model1.layers, ensemble_model.layers)):
        if len(layer1.weights) > 0:  # Skip layers without weights
            # Get weights from original models
            weights1 = layer1.get_weights()
            weights2 = model2.layers[i].get_weights()
            
            # Verify that the weight shapes match
            if [w.shape for w in weights1] != [w.shape for w in weights2]:
                raise ValueError(f"Weight shapes in layer {i} don't match between models")
            
            # Compute average weights
            avg_weights = []
            for w1, w2 in zip(weights1, weights2):
                avg_weights.append((w1 + w2) / 2.0)
            
            # Set the averaged weights to the ensemble model
            layer2.set_weights(avg_weights)
    
    
    
    return ensemble_model

# Example usage
# Assuming lstm_model_1 and lstm_model_2 are your pre-trained LSTM models
# ensemble_lstm = create_lstm_ensemble_with_averaged_weights(lstm_model_1, lstm_model_2)

# To evaluate the ensemble model
# ensemble_lstm.evaluate(test_data, test_labels)

# To make predictions with the ensemble model
# predictions = ensemble_lstm.predict(new_data)
```
