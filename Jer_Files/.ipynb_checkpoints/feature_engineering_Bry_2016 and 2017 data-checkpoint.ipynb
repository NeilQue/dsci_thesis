### Making a rating curve to get hourly data

```python
sto_nino_daily_ave = sto_nino_df.groupby('Date').mean()
sto_nino_daily_ave = sto_nino_daily_ave.reset_index()
sto_nino_daily_ave
```

```python
sto_nino_daily_ave.info()
```

```python
sto_nino_rating_curve = sto_nino_daily_ave.merge(discharge_df[['Discharge_Sto_Nino', 'Date']], on='Date')
sto_nino_rating_curve
```

```python
sto_nino_rating_curve.info()
```

```python
sto_nino_rating_curve_train = sto_nino_rating_curve.dropna()
sto_nino_rating_curve_train
```

```python
sto_nino_rating_curve_train.info()
```

```python
import numpy as np

X = np.array(sto_nino_rating_curve_train['Waterlevel_Sto_Nino'])
Y = np.array(sto_nino_rating_curve_train['Discharge_Sto_Nino'])
```

### Using logarithmic regression

```python
sto_nino_log_model = np.polyfit(np.log(X), Y, 1)
sto_nino_log_model
```

```python
coef = sto_nino_log_model[0]
cons = sto_nino_log_model[1]
```

```python
import matplotlib.pyplot as plt

plt.scatter(sto_nino_rating_curve_train['Waterlevel_Sto_Nino'], sto_nino_rating_curve_train['Discharge_Sto_Nino'])
plt.scatter(sto_nino_rating_curve_train['Waterlevel_Sto_Nino'], cons+coef*np.log(sto_nino_rating_curve_train['Waterlevel_Sto_Nino']))
plt.show()
```

```python
montalban_daily_ave = montalban_df.groupby('Date').mean()
montalban_daily_ave = montalban_daily_ave.reset_index()
montalban_daily_ave
```

```python
montalban_rating_curve = montalban_daily_ave.merge(discharge_df[['Discharge_San_Jose', 'Date']], on='Date')
montalban_rating_curve
```

```python
X2 = np.array(montalban_rating_curve['Waterlevel_Montalban'])
Y2 = np.array(montalban_rating_curve['Discharge_San_Jose'])
```

```python
montalban_log_model = np.polyfit(np.log(X2), Y2, 1)
montalban_log_model
```

```python
coef2 = montalban_log_model[0]
cons2 = montalban_log_model[1]
```

```python
plt.scatter(montalban_rating_curve['Waterlevel_Montalban'], montalban_rating_curve['Discharge_San_Jose'])
plt.scatter(montalban_rating_curve['Waterlevel_Montalban'], cons2+coef2*np.log(montalban_rating_curve['Waterlevel_Montalban']))
plt.show()
```

## Solving for Discharge

```python
final_df['Discharge_Sto_Nino'] = coef*np.log(final_df['Waterlevel_Sto_Nino'])+cons
final_df
```

```python
final_df['Discharge_San_Jose'] = coef2*np.log(final_df['Waterlevel_Montalban'])+cons2
final_df
```

Sanity Checks: making sure that Q will not be less than 0

```python
coef2*np.log(min(final_df['Waterlevel_Montalban']))+cons2 > 0
```

```python
np.log(min(final_df['Waterlevel_Sto_Nino']))*coef + cons > 0
```

## Solving for Cross Sectional Area

```python
final_df['Cross_Section_Sto_Nino'] = 66*final_df['Waterlevel_Sto_Nino']
final_df
```

```python
final_df['Cross_Section_Montalban'] = 30*final_df['Waterlevel_Montalban']
final_df
```

## Solving for Velocity

```python
final_df['Velocity_Sto_Nino'] = final_df['Discharge_Sto_Nino'] / final_df['Cross_Section_Sto_Nino']
final_df['Velocity_Montalban'] = final_df['Discharge_San_Jose'] / final_df['Cross_Section_Montalban']
final_df
```

```python
final_df = final_df.sort_values(by=['Date', 'Time'])
final_df
```

```python
final_df.info()
```

```python
final_df['datetime'] = pd.to_datetime(final_df['Date'].astype(str) + ' ' + final_df['Time'].astype(str))
final_df = final_df.iloc[:, 2:]
final_df
```

```python
final_df = final_df.reset_index()
```

```python
final_df['t'] = pd.Series([(final_df.iloc[i]['datetime'] - final_df.iloc[0]['datetime']).total_seconds() for i in range(len(final_df))])
final_df
```
