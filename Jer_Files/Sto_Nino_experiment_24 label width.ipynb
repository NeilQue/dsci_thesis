```python
import numpy as np
import tensorflow as tf

class WindowGenerator():
    def __init__(self, input_width, label_width, shift, train_df, val_df, test_df, label_columns=None):
        # Store data
        self.train_df = train_df
        self.val_df = val_df
        self.test_df = test_df

        # Label column indices
        self.label_columns = label_columns
        if label_columns is not None:
            self.label_columns_indices = {name: i for i, name in enumerate(label_columns)}
        self.column_indices = {name: i for i, name in enumerate(train_df.columns)}

        # Window parameters
        self.input_width = input_width
        self.label_width = label_width
        self.shift = shift

        self.total_window_size = input_width + shift
        self.input_slice = slice(0, input_width)
        self.input_indices = np.arange(self.total_window_size)[self.input_slice]

        self.label_start = self.total_window_size - self.label_width
        self.labels_slice = slice(self.label_start, None)
        self.label_indices = np.arange(self.total_window_size)[self.labels_slice]

    def __repr__(self):
        return '\n'.join([
            f'Total window size: {self.total_window_size}',
            f'Input indices: {self.input_indices}',
            f'Label indices: {self.label_indices}',
            f'Label column name(s): {self.label_columns}'
        ])

    def split_window(self, features):
        """Split into input and label windows"""
        inputs = features[:, self.input_slice, :]
        labels = features[:, self.labels_slice, :]

        if self.label_columns is not None:
            labels = tf.stack([labels[:, :, self.column_indices[name]] for name in self.label_columns], axis=-1)

        inputs.set_shape([None, self.input_width, None])
        labels.set_shape([None, self.label_width, None])

        return inputs, labels

    def make_dataset(self, data, shuffle=False):
        """Creates tf dataset without shuffling by default"""
        data = np.array(data, dtype=np.float32)
        ds = tf.keras.utils.timeseries_dataset_from_array(
            data=data,
            targets=None,
            sequence_length=self.total_window_size,
            sequence_stride=1,
            shuffle=shuffle,
            batch_size=32,
        )

        ds = ds.map(self.split_window)

        return ds


    @property
    def train(self):
        return self.make_dataset(self.train_df, shuffle=False)  # Ensure training is sequential

    @property
    def val(self):
        return self.make_dataset(self.val_df, shuffle=False)

    @property
    def test(self):
        return self.make_dataset(self.test_df, shuffle=False)

    @property
    def example(self):
        """Get and cache an example batch of `inputs, labels` for plotting."""
        result = getattr(self, '_example', None)
        if result is None:
            result = next(iter(self.train))
            self._example = result
        return result
```

```python
from tensorflow.keras import layers, backend as K

# Nash-Sutcliffe Efficiency function
def NSE(y_true, y_pred):
    y_pred = K.flatten(y_pred)
    y_true = K.flatten(y_true)

    SS_res = K.sum(K.square(y_true - y_pred))
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))

    return 1 - SS_res / (SS_tot + K.epsilon())

    
# Number of input features (e.g., water level + other features)
num_features = len(train_df_sto_nino.columns)


lstm_model_sto_nino = tf.keras.models.Sequential([
    layers.LSTM(64, return_sequences=True, input_shape=(24,train_df_sto_nino.shape[1])),
    layers.Flatten(),
    layers.Dense(units=64, activation='relu'),
    layers.Dense(units=64, activation='relu'),
    layers.Dense(units=1)
])

lstm_model_sto_nino.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=[NSE]
)
```

```python
# Create window object with longer input sequences
window_sto_nino = WindowGenerator(
    input_width=24,  # Using 24 past timesteps for better sequential context
    label_width=1, 
    shift=1,
    train_df=train_df_sto_nino,
    val_df=val_df_sto_nino,
    test_df=test_df_sto_nino,
    label_columns=['Waterlevel']
)

# Early stopping to prevent overfitting
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train model
history = lstm_model_sto_nino.fit(
    window_sto_nino.train,
    validation_data=window_sto_nino.val,
    epochs=100,
    batch_size=32,  # Keep batch size consistent with stateful LSTM
    shuffle=False,  # DO NOT SHUFFLE!
    callbacks=[early_stopping]
)
```

```python
# Evaluate on sequential data (DO NOT shuffle test set)
performance = lstm_model_sto_nino.evaluate(window_sto_nino.test)
print(f"NSE on Sequential Test Data: {performance[1]:.4f}")
```

```python
def rolling_forecast(model, dataset, steps=10):
    """Generates rolling forecasts step by step"""
    predictions = []
    for inputs, labels in dataset:
        pred = model.predict(inputs)  # Predict current timestep
        predictions.append(pred)
    return np.concatenate(predictions, axis=0)

test_preds = rolling_forecast(lstm_model_sto_nino, window_sto_nino.test)
```
