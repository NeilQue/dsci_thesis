### Making a rating curve to get hourly data

```python
sto_nino_daily_ave = sto_nino_df.groupby('Date').mean()
sto_nino_daily_ave = sto_nino_daily_ave.reset_index()
sto_nino_daily_ave
```

```python
sto_nino_daily_ave.info()
```

```python
sto_nino_rating_curve = sto_nino_daily_ave.merge(discharge_df[['Discharge_Sto_Nino', 'Date']], on='Date')
sto_nino_rating_curve
```

```python
sto_nino_rating_curve.info()
```

```python
sto_nino_rating_curve_train = sto_nino_rating_curve.dropna()
sto_nino_rating_curve_train
```

```python
sto_nino_rating_curve_train.info()
```

```python
import numpy as np

X = np.array(sto_nino_rating_curve_train['Discharge_Sto_Nino'])
Y = np.array(sto_nino_rating_curve_train['Waterlevel_Sto_Nino'])
```

```python
min_x = min(X)
max_x = max(X)
print(min_x, max_x)
```

### Using Nonlinear Least Squares (Logarithmic Regression)

```python
from scipy.optimize import curve_fit
def func(x, a, b):
    return a * np.log(x) + b
```

```python
popt, pcov = curve_fit(func, X, Y)

# Get the optimal parameters
print(popt)
a = popt[0]
b = popt[1]
```

Best fitting line is y = 0.4774044log(x) + 11.03946461

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, label = "Data")  
plt.scatter(X, a * np.log(X) + b, label = "Rating Curve")
plt.xlabel('Water Discharge')
plt.ylabel('Water Level')
plt.title('Rating Curve for Sto. Nino')
plt.legend()
plt.show()
```

## Causality between Rainfall and Water Level


Granger causality is a statistical hypothesis test to determine whether one time series can predict another. 
Essentially, it assesses whether past values of one variable contain useful information for predicting future values of another variable. <br>

Hypothesis Test:

H_0: The rainfall time series does not Granger-cause the water level time series. <br>
H_A: The rainfall time series Granger-causes the water level time series.
