## Getting entries from typhoon dates

```python
def get_typhoon_event(start_date, end_date, dataframe=df):
    '''
    Get typhoon event from dataframe
    Input the typhoon start date and end date
    Dates will obtained from:
    https://pubfiles.pagasa.dost.gov.ph/pagasaweb/files/tamss/weather/tcsummary/PAGASA_ARTC_2017.pdf
    https://en.wikipedia.org/wiki/2016_Pacific_typhoon_season#Systems
    '''
    start = dt.datetime.strptime(start_date, '%Y-%m-%d')
    end = dt.datetime.strptime(end_date, '%Y-%m-%d')

    after_start_date = dataframe['datetime'] >= start
    before_end_date = dataframe['datetime'] < end + dt.timedelta(days=1)

    typhoon_event = dataframe[after_start_date & before_end_date]
    
    return typhoon_event
```

```python
def graph_waterlevel(dataframe):
    '''
    Make a graph for water level
    '''
    fig, axes = plt.subplots(2, 1)
    
    axes[0].plot(dataframe['datetime'], dataframe['Waterlevel_Sto_Nino'], color='blue', label='Sto Nino')
    axes[0].set_title('Water Level')
    axes[0].set_ylim(11, 16.5)
    axes[0].set_xticks([])
    axes[0].legend()
    
    axes[1].plot(dataframe['datetime'], dataframe['Waterlevel_Montalban'], color='green', label='Montalban')
    axes[1].legend()
    axes[1].set_ylim(20, 27)
    axes[1].tick_params(axis='x', labelrotation=45)
    plt.show()
```

```python
def reset_t(dataframe):
    dataframe['t'] = np.array(dataframe['t']) - dataframe.iloc[0]['t']
    
    return dataframe
```
