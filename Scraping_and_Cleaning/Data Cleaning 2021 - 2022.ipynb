# Add the 2022 data

```python
df_2022 = pd.read_csv(data_2022)

# Remove the 2023 Jan 1 data
df_2022 = df_2022.iloc[:-26]
df_2022
```

```python
df_2022 = df_2022.loc[(df_2022["station"] == "Mt. Oro") |
            (df_2022["station"] == "Boso Boso") |
            (df_2022["station"] == "Mt. Aries") |
            (df_2022["station"] == "Mt. Campana") |
            (df_2022["station"] == "Nangka")]
```

```python
df_2022 = df_2022[["datetime", "water_level", "station", "1hr"]]
```

```python
df_2022 = df_2022.reset_index()
df_2022 = df_2022.iloc[:, 1:]
```

```python
df_2022
```

```python
pivoted_df = df_2022.pivot_table(index=['datetime', 'water_level'], columns='station', values='1hr').reset_index()
pivoted_df
```

```python
pivoted_df.info()
```

```python
pivoted_df['datetime'] = pd.to_datetime(pivoted_df['datetime'])
```

```python
pivoted_df
```

```python
df_sorted = pivoted_df.sort_values(by='datetime', ascending=True)
df_sorted
```

```python
df_sorted = df_sorted.iloc[:, 1:]
```

# Combine the 2021 and 2022 data

```python
df_sorted = df_sorted.rename(columns={'water_level': 'Waterlevel_Sto_Nino',
                        'Boso Boso': 'Rainfall_Boso',
                        'Mt. Aries': 'Rainfall_Aries',
                        'Mt. Campana': 'Rainfall_Campana',
                        'Mt. Oro': 'Rainfall_Oro',
                        'Nangka': 'Rainfall_Nangka'})
```

```python
df_sorted
```

```python
# Get the intersection of column names between df1 and df2
common_columns = final_df.columns.intersection(df_sorted.columns)

common_columns
```

```python
# Concatenate the DataFrames based on common columns
df_combined = pd.concat([final_df[common_columns], df_sorted[common_columns]], ignore_index=True, join='inner')

# Display the combined DataFrame
df_combined
```

```python
# Export the dataframe
df_combined.to_csv('2021-2022 data.csv')
```
